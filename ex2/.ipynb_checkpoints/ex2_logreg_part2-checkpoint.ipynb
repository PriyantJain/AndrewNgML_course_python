```python
data = np.genfromtxt('ex2data2.txt',delimiter = ',')
m=len(data)
```

```python
X = np.hstack((np.ones((m,1)),data[:,:-1]))
y = data[:,-1].reshape(-1,1)
```

```python
#plotting the data
plt.figure()
yf =y.flatten()
plt.plot(X[yf==0][:,2],X[yf==0][:,1],'bo',label = 'y=0')
plt.plot(X[yf==1][:,2],X[yf==1][:,1],'r+',label = 'y=1')
```

```python
def sigmoid(z):
    return 1/(1+np.e**-z)
```

```python
# it also manages extra line of ones
def addPolyfeatures(X1,X2):
    degree = 6
    k=0
    col_len = round((degree+2)*(degree+1)/2)
    ret = np.ones((m,col_len))
    for i in range(degree+1):
        for j in range(i+1):
            ret[:,k] = (X1**(i-j))*(X2**j)
            k+=1
    return ret
```

```python
def CostGradreg(initial_theta, X, y, lamda=0):
    initial_theta = initial_theta.reshape((-1,1))
    y = y.reshape((-1,1))
    #print(np.shape(X) , (initial_theta))
    h = sigmoid(X@initial_theta)
    thetareg = initial_theta.copy()
    thetareg[0]=0
    J = (-1/m)*(y.T @ np.log(h) + (1-y).T @np.log(1-h) - (lamda/2)*(thetareg.T @ thetareg))
    grad = (1/m) * (X.T @(h-y) + lamda * thetareg)
    return J.item(),grad
```

```python
X_poly = addPolyfeatures(X[:,1],X[:,2])
initial_theta = np.zeros((X_poly.shape[1],1))
```

```python
#testing

#sigmoid(np.array([1,2]))
#CostGradreg(np.array([-24,0.2,0.2]),X,y,0)
#((11.776595295633163,array([[-0.49152542],[-0.00860136],[-0.09147301]]).T)
#CostGradreg(np.ones((X_poly.shape[1],1)),X_poly,y,10)
#(3.16450933161501,array([[0.34604507],[0.16135192],[0.19479576],[0.22686278]..]).T)
```

```python
lamda = 1
options = {'GradObj':'on','maxiter':400}
```

```python
fmincgcost = lambda initheta : CostGradreg(initheta,X_poly,y,lamda)
```

```python
theta, Jrec, *_ = fmincg(fmincgcost,initial_theta,options)
```

```python
#accuracy on our training set (with lambda = 1): 83.051
np.mean(((X_poly@theta)>=0) == y)
```
