```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from fmincg import fmincg
np.random.seed(0)
```

```python
input_layer_size  = 400;  # 20x20 Input Images of Digits
hidden_layer_size = 25;   # 25 hidden units
num_labels = 10;          # 10 labels, from 1 to 10   
```

# Importing data and weights

```python
data = loadmat('ex4data1.mat')
X = data['X']
y = data['y']
m,n = X.shape
```

```python
data = loadmat('ex4weights.mat')
Theta1 = data['Theta1']
Theta2 = data['Theta2']
nn_params = np.vstack((Theta1.T.reshape(-1,1) , Theta2.T.reshape(-1,1)))
```

```python
def displayk2(X,k=10):
    '''function to display numbers data works different than course'''
    m,n = X.shape
    side = int(np.sqrt(n))
    X = X.reshape(k,k*side,side).transpose(0,2,1).reshape(k*side,k*side)
    plt.figure()
    plt.imshow(X , cmap = 'Greys')
    
def sigmoid(z):
    return 1/(1+np.e**-z)

def nnCostFunction(nn_params, input_layer_size, hidden_layer_size,num_labels, X, y, lamda):
    ''' cost function for neural network'''
    #working
    Theta1 = (nn_params[:hidden_layer_size*(input_layer_size+1)]
              .reshape(input_layer_size+1, hidden_layer_size)).T.copy()
    Theta2 = (nn_params[hidden_layer_size*(input_layer_size+1) :]
              .reshape(hidden_layer_size+1, num_labels)).T.copy()
    y = y.reshape((1,-1)) -1
    m,n = X.shape
    a1 = np.hstack((np.ones((m,1)) , X))
    a2 = np.hstack((np.ones((m,1)) , sigmoid(a1 @ Theta1.T)))
    a3 = sigmoid(a2 @ Theta2.T)
    
    Yexp = np.zeros(a3.shape,bool)
    Yexp[np.arange(m),y] = 1
    
    theta_reg = np.hstack((Theta1[:,1:].ravel() , Theta2[:,1:].ravel()))
    J = (-1/m)*(np.sum(np.log(a3[Yexp]))
                + np.sum(np.log(1-(a3[Yexp==False])))
                - (lamda/2)*((theta_reg @ theta_reg.T).item()))
    
    d3 = a3-Yexp
    d2 = (d3 @ Theta2[:,1:] ) * a2[:,1:] * (1-a2[:,1:])
    Theta1[:,0]=0
    Theta2[:,0]=0
    
    Theta1_grad = (d2.T @ a1) + lamda * (Theta1)
    Theta2_grad = (d3.T @ a2) + lamda * (Theta2)
    
    grad = np.vstack((Theta1_grad.T.reshape(-1,1) , Theta2_grad.T.reshape(-1,1)))/m
    
    return J,grad

def randInitializeWeights(L_in, L_out,epsilon = 0.12):
    ''' initializes weight to random values'''
    return (np.random.rand(L_out,L_in+1) *epsilon*2)-epsilon

def fixInitializeWeights(L_in, L_out,epsilon = 0.1):
    '''initializes weight to fixed sin values for checking and debugging purposes'''
    return (np.sin(np.arange(1,(L_out * (L_in+1)) + 1).reshape(L_in+1, L_out)).T)*epsilon

def NumGradFunction(J , theta,epsilon = 0.0001):
    ''' numerically calculates gradient, used to verify back propagation'''
    grad = np.zeros((len(theta),1))
    base,*_ = J(theta)
    for i in range(len(theta)):
        theta[i] += epsilon
        grad[i],*_ = J(theta)
        theta[i] -= epsilon
    
    grad = (grad-base)/epsilon
    return grad

def predict(Theta1,Theta2,X_inp):
    ''' prediction function for neural network'''
    m,n = X_inp.shape
    X = np.hstack((np.ones((m,1)),X_inp))
    a2 = np.vstack((np.ones((1,m)),sigmoid(Theta1 @ X.T)))
    return np.argmax(sigmoid(Theta2 @ a2),0).reshape(-1,1)+1

def checkNNGradients(lamda=0):
    ''' compares  numerical and backpropagation gradients'''
    input_layer_size = 3;
    hidden_layer_size = 5;
    num_labels = 3;
    m = 5;
    Theta1 = fixInitializeWeights(input_layer_size,hidden_layer_size)
    Theta2 = fixInitializeWeights(hidden_layer_size,num_labels)
    X = fixInitializeWeights(input_layer_size-1,m)
    y = 1+(np.arange(1,m+1).reshape(-1,1) % num_labels)
    nn_params = np.vstack((Theta1.T.reshape(-1,1) , Theta2.T.reshape(-1,1)))
    #print(Theta1, Theta2)
    cost1,grad1 = nnCostFunction(nn_params, input_layer_size, hidden_layer_size,num_labels, X, y, lamda)
    numgrad = NumGradFunction(lambda x : nnCostFunction(x,input_layer_size, hidden_layer_size,num_labels, X, y, lamda) 
                              , nn_params)
    print(np.array(tuple(zip(grad1.ravel(),numgrad.ravel()))))
```

```python
init_Theta1 = randInitializeWeights(input_layer_size, hidden_layer_size)
init_Theta2 = randInitializeWeights(hidden_layer_size, num_labels)
init_nn_params = np.vstack((init_Theta1.T.reshape(-1,1) , init_Theta2.T.reshape(-1,1)))

options = {'maxiter':50}
lamda = 1
nn_params, Jrec , *_ =fmincg(lambda x : nnCostFunction(x,input_layer_size, hidden_layer_size,num_labels, X, y, lamda), 
                             init_nn_params , options)
```

```python
Theta1_opt = (nn_params[:hidden_layer_size*(input_layer_size+1)]
              .reshape(input_layer_size+1, hidden_layer_size)).T
Theta2_opt = (nn_params[hidden_layer_size*(input_layer_size+1) :]
              .reshape(hidden_layer_size+1, num_labels)).T
```
