# Importing and visualizing data 

```python
data=np.genfromtxt('ex1data1.txt',delimiter=',')
```

```python
X,y=data[:,:-1].reshape(-1,1),data[:,-1].reshape(-1,1)
```

```python
m=len(X)
m
```

```python
plt.figure(1)
plt.plot(X,y,'o')
```

```python
alpha =0.01
theta=np.array([0,0]).reshape(-1,1)
iterations=1500
```

```python
X=np.hstack((np.ones(m).reshape(-1,1) , X))  #adding ones to X matrix
```

### cost function

```python
def findcost(X,y,theta):
    er= X@theta - y
    return ((er.T @ er)/(2*len(X))).item()
```

```python
findcost(X,y,np.array([-1,2]).reshape(-1,1))
#54.24245508201238 value as ans for theta = np.array([-1,2]).reshape(-1,1)
```

### Gradient descent

```python
def GradDesc(X,y,theta,alpha,iterations):
    Jrec = np.zeros(iterations)
    theta = theta.reshape(-1,1)
    for i in range(iterations):
        er= X@theta -y
        theta = theta - (alpha/m)*( X.T @ er)
        Jrec[i] = ((er.T @ er)/(2*m)).item()
    return theta,Jrec
```

```python
theta_learned, Jrec = GradDesc(X,y,theta,alpha,iterations)
```

```python
theta_learned
```

```python
### plotting desicion boundary and changes in cost with iterations ####
plt.figure(2, figsize = (10,5))

plt.subplot(121)
plt.plot(X[:,1],y,'o')
plt.plot(X[:,1],X@theta_learned,'b-',label='desicion boundary')
plt.legend()
plt.title('Linear regression')

plt.subplot(122)
plt.plot(range(1,iterations),(Jrec.T)[1:])
plt.title('Record of cost')
plt.xlabel('no. of iterations')
plt.ylabel('cost')
```
